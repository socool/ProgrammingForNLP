```python
import re

f = open("data/niraj-muang-klaeng.txt", "r")
result = []
for text in f:
      text_clean = ' '.join(text.split(" ")).split()
      if(text_clean != "" and len(text_clean) > 0 and text_clean != "๏"):
        result.append(text_clean)
        
```

```python
from pythainlp.corpus.common import thai_words
from pythainlp.tokenize import dict_trie

def count_words(word_list):
    count = 0
    #Read File
    #Loop Over Each Line
    #Loop Over Each Word
        #Update Count
    new_words = {"วาศนา","สุราไลย","รดา","ชลา","นัครา","กระแสชล","กอกก","คเน","มากมาย"
                 ,"อาไลย","คนอง","กระเทียม","สท้าน","ลออง","ครึก","สกิด","ภัคตร์","เขตร",
                "แขง","ไอสวรรย์","พงา","อุส่าห์"}
    words = new_words.union(thai_words())
    custom_dictionary_trie = dict_trie(words)

    for word in word_list:
        tokenize = word_tokenize(word,custom_dict=custom_dictionary_trie, engine="longest")
        count += len(tokenize)
    return count
```

```python
count_words(tokenize)
```
