```python
def generate_bigram(words):
    bigram_list = []
    for i in range(len(words)):
        if (i+1) < len(words):
            bigram = words[i]+"---"+words[i+1]
            bigram_list.append(bigram)
    return bigram_list
```

```python
words = lyrics.split(' ')
```

```python
cleaned_words = []
```

```python
for word in words:
    if word.strip().isalpha():
        cleaned_words.append(word.strip())
```

```python
bigrams = generate_bigram(cleaned_words)
```
