```python
from pythainlp.corpus.common import thai_words
from pythainlp.tokenize import dict_trie
from pythainlp import word_tokenize, Tokenizer
```

```python
def count_words():
    data_file = open('data/niraj-muang-klaeng-cleaned.txt')
    new_words = {"วาศนา","สุราไลย","รดา","ชลา","นัครา","กระแสชล","กอกก","คเน","มากมาย"
             ,"อาไลย","คนอง","กระเทียม","สท้าน","ลออง","ครึก","สกิด","ภัคตร์","เขตร",
            "แขง","ไอสวรรย์","พงา","อุส่าห์","พนาไลย","พนาวัน","ศรีมหาราชา","มหาชลาไหล"
            ,"มหาระนพ","ตวัน","บรรไลย"}
    count = 0
    words = new_words.union(thai_words())
    custom_dictionary_trie = dict_trie(words)
    word_to_count = {}
    # for each line
    for poem_line in data_file:
        tokenize = word_tokenize(poem_line.strip(),custom_dict=custom_dictionary_trie, engine="longest")
        for word_index in range(len(tokenize)):
            if tokenize[word_index] not in word_to_count:
                word_to_count[tokenize[word_index]] = 1
            else:
                word_to_count[tokenize[word_index]] = word_to_count[tokenize[word_index]] + 1
            count += 1
    word_count_pair_list = __builtins__.list(word_to_count.items())
    word_count_pair_list.sort(key=lambda x: x[1],reverse=True)
    print(word_count_pair_list)
    output_file = open('data/word_counts.txt','w')
    for pair in word_count_pair_list:
        str_to_print = '{}\t{}\n'.format(pair[0],pair[1])
        output_file.write(str_to_print)
    output_file.close()
    
    return count
        
```

```python
print(count_words())
```
